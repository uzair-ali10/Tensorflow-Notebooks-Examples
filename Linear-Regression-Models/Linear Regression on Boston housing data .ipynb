```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
```

```python
boston=datasets.load_boston()
x=boston.data
y=boston.target
x
```

```python
df=pd.DataFrame(x)
df.columns=boston.feature_names
df.describe()
```

```python
from sklearn import model_selection
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y)
xtrain.shape
```

```python
from sklearn.linear_model import LinearRegression
algo=LinearRegression()
algo.fit(xtrain,ytrain)
```

```python
ypredict=algo.predict(xtest)
```

```python
plt.scatter(ypredict,ytest)
plt.axis([0,40,0,40])
```

```python
plt.plot(ypredict,ytest)
```

```python
score_test=algo.score(xtest,ytest)
score_test
```

```python
score_train=algo.score(xtrain,ytrain)
score_train
```

Example Data Scnerio:
* An Experiment deug was tested on individuals from ages 13 to 100 in a clinical trial.
* The trial had 2100 participants. Half were under 65 years old, half were 65 years or older
* Around 95% of paitents who were 65 or older experinced side effects.
* Around 95% of paitents under 65 experinced no side effects.

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import shuffle
import tensorflow as tf
```

Generating Experimental Dataset

```python
train_samples=[]
train_labels=[]
```

```python
for i in range(50):
    # The ~5% of younger individuals who did experinced side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)

    # The ~5% of older individuals who did no experincved side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    # The ~95% of younger individuals who did experinced side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

    # The ~5% of older individuals who did no experincved side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)   
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

Convert train_labels and train_samples into numpy array

```python
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

Normalizing The Dataset

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

Build the Model

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(units='16',input_shape=(1,), activation='relu'),
    tf.keras.layers.Dense(units='32', activation='relu'),
    tf.keras.layers.Dense(units= 2, activation='softmax')
])
```

```python
model.summary()
```

Complie and Train The model

```python
model.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
x_shuffled,y_shuffled=shuffle(scaled_train_samples,train_labels,random_state=0)
```

```python
model.fit(x=scaled_train_samples,y=train_labels,epochs=30, validation_split=0.1)
```
